```python
from astropy.io import fits
# this is simple code to extract the mjd from a fits file header
```

```python
def mjd_search(filename):
    f_open = fits.open(filename)
    mjd_start = f_open[0].header['expstart']
    mjd_end = f_open[0].header['expend']
    print((mjd_start + mjd_end)/2)
    f_open.close()
```

```python
mjd_search("ic1376010_drc.fits")
```

```python
# Drizzled image input finder + # of images + Date (in yy-dd-mm) + mjd + exptime

def input_search(filename):
    f_open = fits.open(filename)
    i_put = f_open[0].header['input']
    date = f_open[0].header["date-obs"]
    img_num = f_open[0].header["ninput"]
    mjd = f_open[0].header['mjd-obs']
    exptime = f_open[0].header['exptime']
    print(i_put)
    print(date)
    print(img_num)
    print(mjd)
    print(exptime)
    f_open.close()
```

```python
input_search('wfc3.f350lp.ref_0001.drz.fits')
```
